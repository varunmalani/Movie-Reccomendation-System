# Movie Recommendation System

```python
import pandas as pd
import numpy as np
# Converts string into list
import ast
```

```python
movies = pd.read_excel('movies1.xlsx')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
credits.head(2)
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head(2)
```

```python
# THE COLUMNS THAT NEEDS TO BE KEPT
# genres
# id
# keywords
# overview
# title
# cast
# crew
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'Platform', 'cast', 'crew']]
```

```python
movies.head(2)
```

```python
# Counting all the null values in each column
movies.isnull().sum()
```

```python
# Dropping all the null values as the the null value is negligible
movies.dropna(inplace=True)
```

```python
# To check the duplicate values if present or not
movies.duplicated().sum()
```

```python
# This is present in a wierd format (dictionary inside a dictionary)
movies.iloc[0].genres
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj) :
        L.append(i['name'])
    return L
    
```

```python
movies.genres = movies.genres.apply(convert)
```

```python
movies.head(2)
```

```python
movies.keywords = movies.keywords.apply(convert)
```

```python
movies.head(2)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj) :
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
    
```

```python
movies.cast = movies.cast.apply(convert3)
```

```python
movies.head(2)
```

```python
def fetch_director(obj):
    L = []
    for j in ast.literal_eval(obj):
        if (j['job'] == 'Director'):
            L.append(j['name'])
            break
    return L
```

```python
movies.crew = movies.crew.apply(fetch_director)
```

```python
# Converting the string into list
movies.overview = movies.overview.apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
# Removing spaces and creating list comprehension
movies.genres = movies.genres.apply(lambda x:[i.replace(" ","") for i in x])
movies.keywords = movies.keywords.apply(lambda x:[i.replace(" ","") for i in x])
movies.crew = movies.crew.apply(lambda x:[i.replace(" ","") for i in x])
movies.cast = movies.cast.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
movies['tags'] = movies.overview + movies.genres + movies.keywords + movies.cast + movies.crew
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags', 'Platform']]
new_df.sample(5)
```

```python
# Converting a list into string 
new_df.tags = new_df.tags.apply(lambda x: " ".join(x))
```

```python
new_df.tags[0]
```

```python
new_df.tags = new_df.tags.apply(lambda x:x.lower())
```

```python
new_df.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df.tags).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
# Now we are going to apply stemming which removes the similar words
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df.tags = new_df.tags.apply(stem)
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df.tags).toarray()
cv.get_feature_names()
```

```python
# Here we are going to use Cosine distance
# Cosine distance calculates the anlgle between 2 lines
# Here we don't use Euclidean distance as we are dealing with data in a higher dimension
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# Fetching the index of each movies
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]
    
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title,' - ', new_df.iloc[i[0]].Platform)
```

```python
recommend('Star Trek: Insurrection')
```

```python
recommend('Avatar')
```

```python
recommend('Batman')
```

```python
recommend('Iron Man')
```

```python
recommend('Batman Begins')
```
